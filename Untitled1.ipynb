```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import csv

# ---------------------------
# Upload Audio File
# ---------------------------
uploaded = files.upload()
file_name = next(iter(uploaded))    # Get uploaded filename

# ---------------------------
# Load Audio
# ---------------------------
audio, sr = librosa.load(file_name)

# ---------------------------
# Save Filename and Type to CSV
# ---------------------------
csv_name = "audio_dataset.csv"

# Create CSV file with header only once (if not exists)
try:
    open(csv_name, "x").write("Filename,Format\n")   # 'x' fails if file exists
except:
    pass

# Append new row with filename and format
with open(csv_name, "a", newline='') as f:
    writer = csv.writer(f)
    writer.writerow([file_name, "wav"])

print(f"Saved {file_name} in {csv_name}")

# ---------------------------
# Plot Waveform
# ---------------------------
amplitude = np.abs(audio)

plt.figure(figsize=(14, 5))
plt.plot(np.arange(len(audio)) / sr, amplitude)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform (Amplitude Over Time)")
plt.tight_layout()
plt.show()

# ---------------------------
# Plot Spectrogram
# ---------------------------
n_fft = 2048
hop_length = 512
stft = np.abs(librosa.stft(audio, n_fft=n_fft, hop_length=hop_length))

plt.figure(figsize=(14, 6))
plt.imshow(librosa.amplitude_to_db(stft, ref=np.max),
           aspect='auto',
           origin='lower',
           cmap='turbo')
plt.colorbar(format='%+2.0f dB')
plt.xlabel("Time (s)")
plt.ylabel("Frequency Bin")
plt.title("Spectrogram (Frequency over Time)")
plt.tight_layout()
plt.show()
```
